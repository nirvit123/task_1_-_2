```python
class Solution:
    def defangIPaddr(self, address: str) -> str:
        l=[]
        for i in address:
            if i =='.':
                l.append('[.]')
            else:
                l.append(i)
        n_s=''.join(l)
        return n_s
```

```python
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        
        k=[]
        for i in range(len(nums)):
            l=0
            for j in range (len(nums)):
                if nums[i]>nums[j]:
                    l+=1
            k.append(l)
            
        return k        
```

```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        s=str(n)
        l=list(s)
        n=1
        a=0
        for i in l:
            n*=int(i)
            a+=int(i)
        return n-a
```

```python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        r=0
        for i in range (n):
            r^=((start+(2*i)))
        return r
```
