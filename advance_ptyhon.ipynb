```python
class Solution:
    def runningSum(self, nums: List[int]) -> List[int]:
        new=[]
        d=0
        for i in nums:
            d+=i
            new.append(d)
        return new
```

```python
class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        new=[]
        if len(nums) == 2*n:
            for i in range (n):
                x1=nums[i]
                y1=nums[i+n]
                new.append(x1)
                new.append(y1)
        return new
```

```python
class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        new=[]
        ma=max(candies)
        for i in candies:
            if (i+extraCandies)>= ma:
                new.append(True)
            else:
                new.append(False)
        return new
```
